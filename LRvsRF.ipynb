```python
import pandas as pd
import matplotlib.pyplot as plt

lr_df = pd.read_csv("./results/lr_interpretability_stability.csv")
rf_df = pd.read_csv("./results/rf_interpretability_stability.csv")
```

```python
plt.figure(figsize=(8,5))

# Logistic Regression
plt.plot(
    lr_df["noise_level"], 
    lr_df["spearman_rank_correlation"], 
    marker='o', 
    linestyle='-', 
    label='Logistic Regression'
)

# Random Forest
plt.plot(
    rf_df["noise_level"], 
    rf_df["spearman_rank_correlation"], 
    marker='s', 
    linestyle='--', 
    label='Random Forest'
)

plt.xlabel("Noise Level (Data Quality Degradation)")
plt.ylabel("Spearman Rank Correlation")
plt.title("Interpretability Stability: LR vs RF")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
